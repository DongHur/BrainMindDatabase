```python
import numpy as np
from summa import keywords
```

```python
from googleapiclient.discovery import build
import os.path
import pickle

SAMPLE_SPREADSHEET_ID = '1_hyrK6CMgeH4DXGF_boLJP27hdPfGo9tTt9PjFo-rmA'

def init_sheet():
	sheet = None
	if os.path.exists('token.pickle'):
		creds = None
		with open('token.pickle', 'rb') as token:
			creds = pickle.load(token)
		service = build('sheets', 'v4', credentials=creds)
		# Call the Sheets API
		sheet = service.spreadsheets()
		print("Authenticated.")
	else:
		print("Did Not Login.")
	return sheet

sheet = init_sheet()
```

```python
sheet_name = "'Master List (unscrubbed)'!"
range_names = [
    sheet_name+"E3:E",
]
result = sheet.values().batchGet(
    spreadsheetId=SAMPLE_SPREADSHEET_ID, ranges=range_names).execute()

data = result.get('valueRanges')
Description = data[0]['values']
```

```python
keyward_list = []
for i in range(len(Description)):
    text = keywords.keywords(Description[i][0])
    keyward_list.extend([text])
    print(text)
    print("")
```
